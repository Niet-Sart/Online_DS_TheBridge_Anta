### #1.1

Carga el dataset de casas de California que está en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripción. Deshazte de las filas con nulos.

```python
df_cities= pd.read_csv("./data/california_cities.csv", index_col=0)
df_cities['city'] = df_cities['city'].apply(lambda x: ''.join(map(lambda y: y if y.islower() else " "+y, x)))
df_cities
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# 1

df_top5 = df_cities.sort_values(by="population_total", ascending=False)[:5]

x = df_top5.city.unique()
y = df_top5.population_total / 1000

fig = plt.figure(figsize=(9,9)) 
ax = plt.axes() 
plt.ylim([0,5000])
barras = ax.bar(x, y, color=["blue", "grey", "red", "olive", "brown"], alpha=0.5)

ax.set_title("Ciudades más pobladas de California")
ax.set_xlabel("Ciudades")
ax.set_ylabel("Habitantes (miles)")

for barra, poblacion in zip(barras, y):
    ax.text(barra.get_x() + barra.get_width() / 2, barra.get_height(),
            f'{poblacion:,}', ha='center', va='bottom', fontsize=10)
```

```python
# 2

fig = plt.figure(figsize=(9,9)) 
ax = plt.axes()
ax.hist(df_cities.area_total_km2, bins= 50, alpha=0.7, color= "steelblue", label="Hist Area")

plt.xlabel("Superficie ($km2$)", fontdict={"size": 18, "family": "Bradley Hand itc"})
plt.ylabel("Número de ciudades", fontdict={"size": 18, "family": "Bradley Hand itc"})
plt.legend();
```

```python
# 3

plt.figure(figsize=(9, 9))
plt.scatter(df_cities["area_water_km2"], df_cities["population_total"]/1000,  c="g", alpha=0.5)
plt.xlabel("Área de agua en la ciudad $(km^2)$")
plt.ylabel("Población (miles habitantes)")

outliers = df_cities[df_cities["population_total"] > 800000]
for row, outlier in outliers.iterrows():
    plt.annotate("Outliers", 
                 xy=((outlier["area_water_km2"]), outlier["population_total"]/1000),  
                 xytext=((60,3500)),
                 arrowprops=dict(arrowstyle="->"))
```

```python
# 4

plt.figure(figsize=(9, 9))
plt.boxplot(df_cities.elevation_m,
            vert=False,
            patch_artist=True,
            boxprops={"facecolor" : "r"},
            medianprops={"color" : "k"})
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)");
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 20))

# 1
df_top5 = df_cities.sort_values(by="population_total", ascending=False)[:5]

x = df_top5.city.unique()
y = df_top5.population_total / 1000

ax[0,0].set_ylim([0,5000])
barras = ax[0,0].bar(x, y, color=["blue", "grey", "red", "olive", "brown"], alpha=0.5)

ax[0,0].set_title("Ciudades más pobladas de California")
ax[0,0].set_xlabel("Ciudades")
ax[0,0].set_ylabel("Habitantes (miles)")

for barra, poblacion in zip(barras, y):
    ax[0,0].text(barra.get_x() + barra.get_width() / 2, barra.get_height(),
            f'{poblacion:,}', ha='center', va='bottom', fontsize=10)

# 2
ax[0,1].hist(df_cities.area_total_km2, bins= 50, alpha=0.7, color= "steelblue", label="Hist Area")

ax[0,1].set_xlabel("Superficie ($km2$)", fontdict={"size": 18, "family": "Bradley Hand itc"})
ax[0,1].set_ylabel("Número de ciudades", fontdict={"size": 18, "family": "Bradley Hand itc"})
ax[0,1].legend()

# 3

ax[1,0].scatter(df_cities["area_water_km2"], df_cities["population_total"]/1000,  c="g", alpha=0.5)
ax[1,0].set_xlabel("Área de agua en la ciudad $(km^2)$")
ax[1,0].set_ylabel("Población (miles habitantes)")

outliers = df_cities[df_cities["population_total"] > 800000]
for row, outlier in outliers.iterrows():
    ax[1,0].annotate("Outliers", 
                 xy=((outlier["area_water_km2"]), outlier["population_total"]/1000),  
                 xytext=((60,3500)),
                 arrowprops=dict(arrowstyle="->"))


# 4
ax[1,1].boxplot(df_cities.elevation_m,
            vert=False,
            patch_artist=True,
            boxprops={"facecolor" : "r"},
            medianprops={"color" : "k"})
ax[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
ax[1,1].set_xlabel("Altitud (m)")


plt.show()

```
